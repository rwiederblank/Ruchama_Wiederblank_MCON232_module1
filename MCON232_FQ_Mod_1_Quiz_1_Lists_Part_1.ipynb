### A1) `clean_word`

**Task:** Create a function named `clean_word` that:
- takes in a string `s`
- returns the cleaned version of the string with leading/trailing whitespace removed using `strip()`

**Required function header:**
```python
def clean_word(s: str) -> str:
```


````python
# WRITE YOUR FUNCTION HERE
def clean_word(s: str) -> str:
  cleaned_word = s.strip()
  return cleaned_word

````

````python
# Tests for A1
print(clean_word("  hello  "))   # expected: "hello"
print(clean_word("\nHi\n"))     # expected: "Hi"
````

### A2) `count_letter`

**Task:** Create a function named `count_letter` that:
- takes in a string `text` and a single character `ch`
- returns how many times `ch` appears in `text`




````python
# WRITE YOUR FUNCTION HERE
def count_letter(text, char):
  letter_count = 0
  for letter in text:
    if letter == char:
      letter_count+=1
  return letter_count
````

````python
# Tests for A2
print(count_letter("banana", "a"))  # expected: 3
print(count_letter("banana", "z"))  # expected: 0
````

### A3) `split_name_score`

**Task:** Create a function named `split_name_score` that:
- takes in a string line like `"Alice,90"` or `"Alice,90\n"`
- uses `strip()` then `split(",")`
- returns a tuple: `(name, score_int)`
- - `return (name, score_int)`




````python
# WRITE YOUR FUNCTION HERE
def split_name_score(my_str):
  my_list = my_str.strip().split(",")
  my_tuple = (my_list[0], my_list[1])
  return my_tuple
````

````python
# Tests for A3
print(split_name_score("Alice,90\n"))  # expected: ("Alice", 90)
print(split_name_score("Ben,72"))       # expected: ("Ben", 72)
````

### B1) `count_big`

**Task:** Create a function named `count_big` that:
- takes a list of integers `nums` and an integer `threshold`
- returns how many values in `nums` are **>= threshold**


````python
# WRITE YOUR FUNCTION HERE
def count_big(nums, threshold):
  count = 0
  for num in nums:
    if num >= threshold:
      count += 1
  return count
````

````python
# Tests for B1
print(count_big([3, 10, 11, 2, 10], 10))  # expected: 3
print(count_big([], 10))                  # expected: 0
````

### B2) `make_evens_list`

**Task:** Create a function named `make_evens_list` that:
- takes a list of integers
- returns a **new list** containing only the even numbers (in the same order)





````python
# WRITE YOUR FUNCTION HERE
def make_evens_list(nums):
  evens_list = []
  for num in nums:
    if num%2 == 0:
      evens_list.append(num)
  return evens_list
````

````python
# Tests for B2
print(make_evens_list([1, 2, 3, 4, 5, 6]))  # expected: [2, 4, 6]
print(make_evens_list([1, 3, 5]))           # expected: []
````

### B3) `first_index_of`

**Task:** Create a function named `first_index_of` that:
- takes a list of strings `items` and a string `target`
- returns the index of the first occurrence of `target`
- returns `-1` if not found
- **Do not** use `.index()`


````python
# WRITE YOUR FUNCTION HERE
def first_index_of(items, target):
  for i in range(len(items)):
    if items[i] == target:
      return i
  return -1
````

````python
# Tests for B3
print(first_index_of(["hi", "hello", "hey"], "hello"))  # expected: 1
print(first_index_of(["hi", "hello", "hey"], "bye"))    # expected: -1
````

### B4) `add_bonus`

**Task:** Create a function named `add_bonus` that:
- takes a list of integers `grades` and an integer `bonus`
- returns a **new list** where each grade has `bonus` added




````python
# WRITE YOUR FUNCTION HERE
def add_bonus(grades, bonus):
  updated_grades = []
  for grade in grades:
    updated_grades.append(grade+bonus)
  return updated_grades
````

````python
# Tests for B4
print(add_bonus([70, 80, 90], 5))  # expected: [75, 85, 95]
print(add_bonus([], 5))            # expected: []
````

### C1) `read_grades_file`

**Task:** Create a function named `read_grades_file` that:
- reads `grades.txt`
- returns **two parallel lists**: `(names, scores)`
- uses `strip()` and `split(",")`
- converts the score to `int`




````python
# WRITE YOUR FUNCTION HERE
def read_grades_file(file):
  names = []
  scores = []
  with open(file, "r") as f:
    for line in f:
      line = line.strip().split(",")
      names.append(line[0])
      scores.append(line[1])
  return names, scores

````

````python
# Tests for C1
names, scores = read_grades_file("grades.txt")
print(names)   # expected: ['Alice', 'Ben', 'Cara', 'Dan', 'Eva']
print(scores)  # expected: [90, 72, 88, 95, 72]
````

### C2) `top_student`

**Task:** Create a function named `top_student` that:
- takes two parallel lists: `names` and `scores`
- returns `(top_name, top_score)` for the **highest** score

Assume:
- lists are the same length
- there is at least 1 student




````python
# WRITE YOUR FUNCTION HERE
def top_student(names, scores):
  highest_score = 0
  highest_name = ""
  for index, score in enumerate(scores):
    if index == 0 or score > highest_score:
      highest_score = score
      highest_name = names[index]
  return highest_name, highest_score
````

````python
# Tests for C2
names, scores = read_grades_file("grades.txt")
print(top_student(names, scores))  # expected: ('Dan', 95)
````

### C3) `write_report`

**Task:** Create a function named `write_report` that:
- takes `names`, `scores`, and optional `out_file`
- writes a report file with:
  - Average score (float)
  - Top student line
  - Count of 72
  - Passing names (score >= 75)
- returns `True` when done


````python
# WRITE YOUR FUNCTION HERE
def write_report(names, scores, outfile):
  # finding the average
  avg = 0
  sum = 0
  for score in scores:
    sum += int(score)
  avg = sum / len(scores)

  # top student line
  top_student_name, top_student_grade = top_student(names, scores)

  # count of 72
  over_72 = []
  for i in range(len(scores)):
    if int(scores[i]) >= 72:
      my_tup = (names[i], scores[i])
      over_72.append(my_tup)

  # passing grades
  over_75 = []
  for i in range(len(scores)):
    if int(scores[i]) >= 75:
      my_tup = (names[i], scores[i])
      over_75.append(my_tup)

  # writing it to a file
  with open(outfile, "w") as f:
    f.write(f"The average is {avg}\n")
    f.write(f"The top student is {top_student_name}, scoring, {top_student_grade}.\n")
    f.write(f"Count of 72: {over_72}\n")
    f.write(f"Passing students: {over_75}\n")



````

````python
# Tests for C3
names, scores = read_grades_file("grades.txt")
ok = write_report(names, scores, "report.txt")
print("write_report returned:", ok)

with open("report.txt", "r") as f:
    print("\n--- report.txt ---")
    print(f.read())
````

## Part D â€” Small Thinking Problem (15 minutes)

### D1) `unique_in_order`

**Task:** Create a function named `unique_in_order` that:
- takes a list of integers
- returns a **new** list containing each number only once, in the order it first appears
- does **not** use sets


````python
# WRITE YOUR FUNCTION HERE
def unique_in_order(nums):
  ordered = []
  for num in nums:
    if num not in ordered:
      ordered.append(num)
  return(ordered)


````

````python
# Tests for D1
print(unique_in_order([2, 5, 2, 9, 5, 1]))        # expected: [2, 5, 9, 1]
print(unique_in_order([1, 1, 1, 1]))              # expected: [1]
print(unique_in_order([]))                        # expected: []
````
